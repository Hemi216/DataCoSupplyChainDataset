**Changing 0 and 1 to Delivered as per schedule and  Late Delivery

```python
df["Late_delivery_risk"].unique()
```

```python
def delivery_status(x):
    if x == 1:
        return 'Late Delivery'
    else:
        return 'Delivered as per schedule'

df['Late_delivery_risk'] = df['Late_delivery_risk'].apply(delivery_status)
```
